## 3. Data Loading

````python
def load_credit_default_dataset():
    """
    Load Credit Card Default dataset
    Note: Dataset is in Excel format
    """
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls"
    
    try:
        # Read Excel file, skip first row (extra header)
        df = pd.read_excel(url, header=1)
        print(f"Dataset loaded successfully!")
    except Exception as e:
        print(f"Error loading dataset: {e}")
        print("\nAlternative: Download manually from:")
        print("https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients")
        print("\nThen load with: df = pd.read_excel('default of credit card clients.xls', header=1)")
        return None
    
    print(f"Total samples: {len(df)}")
    print(f"\nColumn names:")
    print(df.columns.tolist())
    
    return df

df = load_credit_default_dataset()
if df is not None:
    df.head()
````

## 4. Exploratory Data Analysis

````python
if df is not None:
    print("Dataset Shape:", df.shape)
    print("\nColumn Info:")
    print(df.info())
    
    # Target column is usually 'default payment next month' or similar
    target_col = [col for col in df.columns if 'default' in col.lower()][0]
    print(f"\nTarget column: '{target_col}'")
    print(f"Target Distribution:")
    print(df[target_col].value_counts())
    
    # Visualize target distribution
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    df[target_col].value_counts().plot(kind='bar', ax=ax[0], color=['#2ecc71', '#e74c3c'])
    ax[0].set_title('Default Distribution (Count)', fontweight='bold')
    ax[0].set_xticklabels(['No Default', 'Default'], rotation=0)
    
    df[target_col].value_counts(normalize=True).plot(kind='bar', ax=ax[1], color=['#2ecc71', '#e74c3c'])
    ax[1].set_title('Default Distribution (Proportion)', fontweight='bold')
    ax[1].set_xticklabels(['No Default', 'Default'], rotation=0)
    
    plt.tight_layout()
    plt.show()
````

## 5. Data Preprocessing

````python
def preprocess_credit_default_data(df):
    """
    Preprocess Credit Card Default dataset
    """
    df = df.copy()
    
    # Find target column
    target_col = [col for col in df.columns if 'default' in col.lower()][0]
    
    # Drop ID column if exists
    if 'ID' in df.columns:
        df = df.drop('ID', axis=1)
    
    # Separate features and target
    X = df.drop(target_col, axis=1)
    y = df[target_col].values
    
    print(f"Features shape: {X.shape}")
    print(f"Class distribution: {np.bincount(y)}")
    
    # Scale all features
    print("\nScaling features...")
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    X_scaled = X_scaled.astype(np.float32)
    y = y.astype(np.int64)
    
    print("Preprocessing complete!")
    return X_scaled, y, scaler

if df is not None:
    X, y, scaler = preprocess_credit_default_data(df)
````
